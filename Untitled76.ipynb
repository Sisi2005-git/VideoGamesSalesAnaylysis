```python
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/VideoGamesSales.csv')
```

```python
df['Region']=df['Region'].fillna('North')
df["NA_Sales"]=df["NA_Sales"].replace('[$]','',regex=True)
df.head()
```

```python
df["NA_Sales"]=pd.to_numeric(df["NA_Sales"], errors="coerce")
```

```python
Average_Sales=df["NA_Sales"].mean()
Average_Sales=int(Average_Sales)
Average_Sales
df["NA_Sales"]=df["NA_Sales"].fillna(Average_Sales)
df.head(20)
```

```python
df["Country"]=df["Country"].replace('USA','United States')
df.head(20)
```

```python
dp=df[df["Country"]=='Australia']
dp.shape[0]
```

```python
df=df.rename(columns={'NA_Sales': 'National Sales','Global_Sales': 'Global Sales','NA_Profit': 'National Profit','Global_Profit':'Global Profit'})
df
```

```python
sales_cap=df["National Sales"].quantile(0.95)
sales_cap=int(sales_cap)
sales_cap
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
National_Sales = df.groupby(['Region','Country'] )['National Sales'].sum().reset_index().sort_values(by = 'National Sales' , ascending=False)
National_Sales
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Region',y='National Sales',hue='Country',data=National_Sales)
plt.title('National Sales by Region and Country')
plt.xlabel('Region')
plt.ylabel('National Sales')
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.boxplot(x='Country',y='National Sales',hue='Region',data=df)
plt.title('Global Sales by Region, Country, and Genre',fontweight='bold')
plt.xlabel('Region',fontweight='bold')
plt.ylabel('Global Sales',fontweight='bold')
plt.show()
```

```python
Sales=df.groupby(['Country'])[['National Sales','Global Sales']].sum().reset_index()
National_Sales= Sales['National Sales']
Country=Sales['Country']
Global_Sales=Sales['Global Sales']
fig, axs = plt.subplots(1,2, figsize=(10,7))

axs[0].pie(National_Sales,labels=Country,autopct='%1.1f%%')
axs[0].set_title('National Sales by Country')

axs[1].pie(Global_Sales,labels=Country,autopct='%1.1f%%')
axs[1].set_title('Global Sales by Country')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Year'], df['National Sales'], marker='o', linestyle='-', color='b', label='National Sales')
plt.plot(df['Year'], df['Global Sales'], marker='s', linestyle='-', color='r', label='Global Sales')
plt.legend()
plt.title('National Sales Over Time')
plt.xlabel('Year')
plt.ylabel('National Sales')
plt.grid(True)
plt.show()
```
